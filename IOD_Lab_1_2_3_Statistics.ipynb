````python
# sample points
X = [0, 5, 10, 15, 20]
Y = [0, 7, 10, 13, 20]

# solve for a and b
def best_fit(X, Y):
    '''
    Return m, b
    # ANSWER
    '''

    m = np.polyfit(X, Y, 1)[0]
    b = np.polyfit(X, Y, 1)[1]
    print(m)
    print(b)

    return m, b

# solution
m, b = best_fit(X, Y)

# plot points and fit line
# ANSWER
plt.scatter(X, Y)
plt.plot(X, [m * x + b for x in X])
plt.show()
````
